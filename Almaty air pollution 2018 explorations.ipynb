```python
import csv
import altair as alt
import pandas as pd

df = pd.read_csv('ap_ala_2018.csv')
del df['Unnamed: 0']
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
alt.data_transformers.disable_max_rows()
chart = alt.Chart(df).mark_trail().encode(
    x='Date:T',
    y=alt.Y('Value:Q', scale = alt.Scale(type='sqrt', domain=(0, 7.5))),
    color='Variable:N',
).configure_view(
    continuousHeight=300,
    continuousWidth=800,
)
chart.interactive()
```

```python
alt.data_transformers.disable_max_rows()
chart = alt.Chart(df).mark_circle().encode(
    x='Date:T',
    y=alt.Y('Value:Q', scale = alt.Scale(type='sqrt', domain=(0, 7.5))),
    color='Variable:N',
).configure_view(
    continuousHeight=300,
    continuousWidth=800,
)
chart.interactive()
```

```python
other_sum = 0
SO2_sum = 0
NO2_sum = 0
CO2_sum = 0
for_sum = 0
fenol_sum = 0

for idx, row in df.iterrows():
    if  df.loc[idx,'Variable'] == 'other':
        other_sum = other_sum + df.loc[idx,'Value']
    
    if  df.loc[idx,'Variable'] == 'SO2':
        SO2_sum = SO2_sum + df.loc[idx,'Value']
        
    if  df.loc[idx,'Variable'] == 'NO2':
        NO2_sum = NO2_sum + df.loc[idx,'Value']
    
    if  df.loc[idx,'Variable'] == 'CO2':
        CO2_sum = CO2_sum + df.loc[idx,'Value']
        
    if  df.loc[idx,'Variable'] == 'formaldehyde':
        for_sum = for_sum + df.loc[idx,'Value']
        
    if  df.loc[idx,'Variable'] == 'fenol':
        fenol_sum = fenol_sum + df.loc[idx,'Value']


d = {'Variable':['PM2.5','SO2', 'NO2', 'CO2', 'formaldehyde', 'fenol' ],
     'Values':[other_sum/365, SO2_sum/365, NO2_sum/365, CO2_sum/365, for_sum/365, fenol_sum/365]}
avg = pd.DataFrame(d)
avg




        
```
